Ex 3: Write a function called number_of_factors that takes an integer and returns how many
factors the number has.

```python
def number_of_factors(n):
    count = 0
    for i in range(1, n + 1):
        if n % i == 0:
            count += 1
    return count
number_of_factors(2)
```

Ex4: Write a function called first_diff that is given two strings and returns the first location in
which the strings differ. If the strings are identical, it should return -1.Ignore the case in your function. It means h and H are the same.

```python
def first_diff(string1, string2):
    s1 = string1.lower()
    s2 = string2.lower()

    min_length = min(len(s1), len(s2))

    for i in range(min_length):
        if s1[i] != s2[i]:
            return i

    if len(s1) != len(s2):
        return min_length

    return -1
first_diff("He", "HELLO")
```

Ex6: Write a function that solves the system of equations ax + by = e and cx +dy = f. Your function should takes 6 parameters, (a, b, c, d, e,f) and return the values of x and y.

```python
def solve_system(a, b, c, d, e, f):
   determinant = a * d - b * c
   x = (e * d - b * f) / determinant
   y = (a * f - e * c) / determinant
   return x, y
solve_system(1, 2, 3, 4, 5, 6)
```

Ex7:  Write a function called change_case that given a string, returns a string with each upper  case letter replaced by a lower case letter and vice-versa

```python
def change_case(text):
  result = ""
  for char in text:
        result += char.swapcase()
  return result
change_case ('Zin Me Me Kyaw')
```

Ex10:  Write a function that is given a 9 × 9 potentially solved Sudoku and returns True if it is solved correctly and False if there is a mistake. The Sudoku is correctly solved if there are no repeated numbers in any row or any column or in any of the nine “blocks.”


```python
def is_sudoku_valid(board):

    def is_valid_group(group):
        return sorted(group) == list(range(1, 10))

    for row in board:
        if not is_valid_group(row):
            return False

    for col in range(9):
        column_data = [board[row][col] for row in range(9)]
        if not is_valid_group(column_data):
            return False

    for r in range(0, 9, 3):
        for c in range(0, 9, 3):
            block = []
            for i in range(3):
                for j in range(3):
                    block.append(board[r + i][c + j])
            if not is_valid_group(block):
                return False

    return True
is_sudoku_valid([
    [1, 0, 2],
    [0, 1, 0],
    [2, 0, 0]
])
```
